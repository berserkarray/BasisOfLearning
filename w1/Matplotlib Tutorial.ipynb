# Matplotlib

It is the plotting library in python that serves as a visualization entity 

The first step is to import <b> Matplotlib </b> library to the current jupyter notebook. Most of its utilities lie under the <b>pyplot</b> submodule.

```python
import matplotlib.pyplot as plt #Shorthand notation is given to facilitate easier further usage
```

```python
import numpy as np

xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints,ypoints) #plots a line
plt.show()
```

```python
#plotting multiple points
xpoints=np.array([1,5,7,8])
ypoints=np.array([2,4,5,6])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#Markers
ypoints=np.array([1,2,5,4,7,6])
plt.plot(ypoints,marker="o",ms=10,mec="magenta",mfc="red") #ms defines marker size, mec defines marker edge color, mfc defines marker face color
plt.show()
```

```python
#linestyle
ypoints=np.array([1,2,5,4,7,6])
plt.plot(ypoints,marker="*",ms=10,linestyle="dotted",color="red",lw=3) #color defines line color, lw defines linewidth
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")    #create lables for the axes of the plot
plt.ylabel("Calorie Burnage")
plt.title("Data")  #give title to the plot
plt.grid()        #add grid to the plot

plt.show()
```

## Subplot function

It is used to draw multiple plots in one figure.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)  
#plt.subplot takes three arguments, first: the number of rows, second: the number of columns, third: the position of the plot
plt.plot(x,y)
plt.title("plot1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("plot2")
 
plt.suptitle("Plots") #add title for the whole figure
plt.show()
```
