## plotting methods

by plot function you create a graph and to display it show function used

\*\*\*\*\*\*line graph\*\*\* plt.plot(\[1,2,3\],\[4,5,6\]) plt.show()

## attributes

\*\*label function is used to create axis names

plt.plot(\[1,2,3\],\[4,5,6\]) plt.xlabel(“jkl”) plt.ylabel(“lkj”)
plt.show()

**title function is used to give graph a name **legend is used in bar
and histogram

plt.title(“regression_demo”) plt.legend()

\*\*\*\*\*\*bar graph label for each bar and color for different graph

plt.bar(\[1,2,3\],\[4,5,6\],label=“pros”,color=‘g’)
plt.bar(\[7,8,9\],\[10,5,8\],label=“os”,color=‘k’) plt.show()

\*\*linewidth is used for differing line width

plt.bar(\[1,2,3\],\[4,5,6\],label=“pros”,color=‘g’,linewidth=10)
plt.bar(\[7,8,9\],\[10,5,8\],label=“os”,color=‘k’,linewidth=14)
plt.show()

\*\*grid is used for developing a grid

plt.bar(\[1,2,3\],\[4,5,6\],label=“pros”,color=‘g’,linewidth=10)
plt.bar(\[7,8,9\],\[10,5,8\],label=“os”,color=‘k’,linewidth=14)
plt.grid(‘true’,color=‘c’) plt.show()
