# Pandas

Pandas is a library that is used for working with datasets. It allows us to analyse big data and make conclusions based on statistical theories.

The first step is to import the <b>Pandas</b> library to the current jupyter notebook.

```python
import pandas as pd #Shorthand notation given to facilitate easier futher usage.
```

<b>Pandas Series</b> : It is a 1 dimensional array holding data of any type.

```python
a=[2,3,4]
col=pd.Series(a)
print(col)

print(col[0]) #returns the value at the index 0
```

```python
#From a dictionary
Heartbeat = {"12PM":80, "1PM":90, "2PM":85}
heartrate_series=pd.Series(Heartbeat)
print(heartrate_series)
```

<b>Pandas Dataframe </b>: It is a multidemsional data structure, like a mutlti dimensional array, or a table with rows and columns.

```python
data={
    "calories":[450,250,500],
    "heartrate":[80,90,85]
}

df = pd.DataFrame(data) #laoding data into a DataFrame object.
print(df)
```

```python
#Locating Rows using loc attribute
print(df.loc[0])
print(df.loc[0:1])
```

```python
#Naming indexes
df2=pd.DataFrame(data,index=["Day1","Day2","Day3"])
print(df2)
```

```python
df=pd.read_csv("dirtydata.csv")
print(df.to_string())
```

```python
new_df=df.dropna() #drops the rows with null values and creates a new DataFrame
print(new_df.to_string())
```

```python
df.dropna(inplace=True) #makes changes to the original DataFrame
print(df.to_string())
```

```python
df=pd.read_csv("data.csv")
print(df.to_string())
```

```python
df.fillna(120,inplace=True) #replaces the null value instead of deleting the entire row
print(df.to_string())
```

```python
df=pd.read_csv("data.csv")
x=df["Calories"].mean()  
df["Calories"].fillna(x,inplace=True) #replaces the null values with the mean of the rest of the column values
print(df.to_string())
#same can be done with median and mode
```

```python
df=pd.read_csv("dirtydata.csv")
print(df.to_string())
```

```python
#the dates in the rows 22 and 26 are in the wrong format
df["Date"] = pd.to_datetime(df["Date"])
print(df.to_string())
```

```python
#the format gets changed but there is NaT generated, to delele the row
df.dropna(subset=["Date"],inplace=True)
print(df.to_string())
```

```python
#replacing wrong values
df.loc[7,"Duration"]=45
print(df.to_string())
```

```python
df=pd.read_csv("data.csv")
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.loc[x,"Duration"]=120  #replacing the values as per a condition
print(df.to_string()) 
```

```python
df=pd.read_csv("Data.csv")
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.drop(x,inplace=True)   #drops the rows which do not satisfy the mentioned criteria
print(df.to_string())
```
