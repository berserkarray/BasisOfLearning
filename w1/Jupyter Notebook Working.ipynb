<b>The Jupyter Notebook App</b> is a server-client application that allows editing and running notebook documents via a web browser. It can be executed on a local desktop requiring no internet access or can be installed on a remote server and accessed through the internet.



<h3>Components of the User-Interface:</h3><br>

<h4>Notebook Dashboard</h4><br>
It is the first window that launches in the browser upon opening the <b>Jupyter Notebook</b> through Anaconda Navigator.

<h4>Notebook Editor</h4><br>
Once we click on a notebook, a notebook editor is opened.


<br>

The Notebook Editor is made up of various cells. When a cell is in <b>Edit mode</b>, the Cell Mode Indicator will change to reflect the cells state. This state is indicated by a small pencil icon on the top right of the interface. When the cell is in <b>Command mode</b>, there is no icon in that location.<br>

The Notebook Editor also various provides controls in the ribbon, to add/delete cells, etc. One of them inlcude the option to choose the type of content to be added to the cell, vis-a-vis, <b>Markdown</b>, <b>Code</b>, <b>RawNBConvert</b>, <b>Heading</b>.<br>
<ul>
    <li><b>Markdown: </b>This is used to add text to the notebook.</li>
    <li><b>Code: </b>This is where the code is written in the language of the kernel</li>
</ul>



<br><br><br>


<h4>Kernel</h4>

A notebook kernel is a computational engine that executes the code contained in a Notebook document. The ipython kernel, referenced here, executes python code. Kernels for many other languages exist (official kernels).
