```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = np.array([0,1,2,3,4])
y = np.array([0,2,4,6,8])

## To set the size of the graph where figsize is propotion and dpi is dots per inch
plt.figure(figsize = (5,3),dpi = 100)


## plots the y-x graph
##plot(x-coorinate,y-coordinates,line's look,label for legend,colour)
a = plt.plot(x,y,'.--',label = '2x',color = 'red')

##x squared graph just because i can
##The graph gets more precise as we shorten the intervals
x2 = np.arange(0,4,0.05)

##for projection like piece
plt.plot(x2[:60],x2[:60]**2,'b',label = 'x**2')
plt.plot(x2[59:],x2[59:]**2,'b--')

## title to the graph 
plt.title('Y vs X',fontdict =  {'fontname':'Comic Sans MS','fontsize':20})

##x and y label
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## to choose the ticks on graph by oneself
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

# to pass the label to the graph
plt.legend()

##to save the graph


## removes the header file showing earlier
plt.show()
```

# Bar Charts

```python
##for x axis
labels = np.array(['A','B','C'])

##for y axis
values = np.array([1,4,2])

##to plot bar graph
bars = plt.bar(labels,values)


##for hashing
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

patterns = ['/','o','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

```python
# To read csv file
gas_prices = pd.read_csv('gas.csv')

plt.figure(figsize = (8,5),dpi = 300)

#for title
plt.title('Gas Prices Across the world',fontdict = {"fontsize" : '15','fontname' : 'Comic Sans MS'})


#to Plot individal line graphs
# plt.plot(gas_prices.Year,gas_prices.USA,'.-',label = 'USA')
# plt.plot(gas_prices.Year,gas_prices.Canada,'.-',label = 'Canada')
# plt.plot(gas_prices['Year'],gas_prices['South Korea'],'.-',label = 'South Korea')
# plt.plot(gas_prices.Year,gas_prices.Australia,'.-')

##to print many values in one go
for country in gas_prices:
    if country !='Year':
        plt.plot(gas_prices.Year,gas_prices[country],'.-',label = country)

# for uniform x-ticks
# print(gas_prices.Year[::3])
plt.xticks(gas_prices.Year[::3].tolist() +[2011])


plt.xlabel('Year',fontdict = {"fontsize" : '15','fontname' : 'Comic Sans MS'})
plt.ylabel('Gas Price(in USD)',fontdict = {"fontsize" : '15','fontname' : 'Comic Sans MS',})


plt.legend()
plt.show()

plt.savefig('Gas_price_fig',dpi = 300)

gas_prices
```

# Histogram 

```python
football = pd.read_csv('football.csv')
```

```python
football.head(5)
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(football.Overall,bins = bins,color = 'red')

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Skills')

plt.show()
```

# Pie Charts

```python
plt.figure(figsize = (5,5))

left = football.loc[football['Preferred Foot'] == 'Left'].count()[0]
right = football.loc[football['Preferred Foot'] == 'Right'].count()[0]
colors = ['red','blue']

# the pie plot code autopct is for percentage
plt.pie([left,right],labels = ['Left','Right'] , colors = colors,autopct = '%.2f')

plt.title('Foot Pref')
#plt.legend()
plt.show()
```

```python
#to remove lbs from weight 
football.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in football.Weight]
football.Weight.head(5)

plt.figure(figsize = (8,5),dpi = 100)

## to create partitions
light = football.loc[football.Weight < 125].count()[0]
light_medium = football.loc[(football.Weight >= 125) & (football.Weight < 150)].count()[0]
medium = football.loc[(football.Weight >= 150) & (football.Weight < 175)].count()[0]
medium_heavy = football.loc[(football.Weight >= 175) & (football.Weight < 200)].count()[0]
heavy = football.loc[football.Weight > 200].count()[0]


weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']

## to explode the graph away from each other
explode = (.4,.1,0,0.1,.4)

plt.title('Weight Distribution of Fifa Players (in lbs)')

plt.pie(weights,labels = labels,autopct = '%.2f',pctdistance = 0.8,explode = explode)

plt.show()
```
