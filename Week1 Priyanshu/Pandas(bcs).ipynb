# Loading data into Pandas

```python
import pandas as pd
import numpy as np
```

```python
#reading from a csv file
poke = pd.read_csv('pokemon.csv')
print(poke)
```

# Reading data in Pandas

```python
# Read Headers
print(poke.columns)
```

```python
# Read each column
print(poke.Name[0:5])
```

```python
# Reading each row
print(poke.iloc[0:4])
```

```python
# To read a specific location
print(poke.iloc[2,1])
```

```python
# for running conditions

poke.loc[poke['Type 1'] == 'Fire']
```

# Sorting/Describing Data

```python
#all stats for the columns
poke.describe()
```

```python
#to sort the table based on one factor
poke.sort_values('Name',ascending =1)
```

```python
#To sort based on multiple factors
poke.sort_values(['Type 1','HP'],ascending =[1,0])
```

# Making changes to data

```python
poke.head(5)
```

```python
# To make a total of all powers
#poke['Total'] = poke['HP'] + poke['Attack']+poke['Defense'] + poke['Sp. Atk'] + poke['Sp. Def']+ poke['Speed']
poke.head(5)

# better method
poke['Total'] = poke.iloc[:,4:10].sum(axis = 1)
poke.head(5)
```

```python
#to delete a column

#poke = poke.drop(columns = ['Total'])
#poke.head(5)
```

```python
# Shifting a column to another place

poke['Total'] = poke.iloc[:,4:10].sum(axis = 1)

cols = list(poke.columns)
poke = poke[cols[0:4] + [cols[-1]] + cols[4:-1]]

poke.head(5)
```

# Saving modified file

```python
poke.to_csv('modified.csv')
```

```python
poke.to_excel('modified1.xlsx',index =False)
```

# Filtering Data

```python
#Bool
#in pandas we use & instead of and

new_x = poke.loc[(poke['Type 1'] == 'Grass') & (poke['Type 2'] == 'Poison')]

new_x.to_csv('filtered.csv')

new_x
```

```python
# to get new indexes for our new table

new_x.reset_index(drop = True,inplace = True)

print(new_x)
```

# Aggregate Statistics(groupby)

```python
#mean
poke.groupby(['Type 1']).mean().sort_values('Total',ascending = 0)
```

```python
#count like this
poke.groupby(['Type 1']).count()

#or

poke['count'] = 1

print(poke.groupby(['Type 1']).count()['count'])

poke = poke.drop(columns = ['count'])
```
