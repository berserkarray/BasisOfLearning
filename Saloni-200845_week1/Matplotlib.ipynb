# Matplotlib

This is a python library

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

### Line Plot

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('My first line plot')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1,color='g',linestyle =':',linewidth = 2)     #adding 2 lines to the same plot, can change the attributes of the lines to show difference
plt.plot(x,y2,color ='r',linestyle = '-')
plt.title("2 line plot")
plt.grid(True)       #adding grid to the background
plt.show()
```

```python
plt.title("2 subplots")
plt.subplot(1,2,1)           #this is like a matrix, 1,2 denote 1 row 2 columns and the last argument 1 represents the index
plt.plot(x,y1,color='g',linestyle =':',linewidth = 2)
plt.subplot(1,2,2)
plt.plot(x,y2,color ='r',linestyle = '-')
plt.grid(True)
plt.show()
```

```python
plt.title("2 subplots")
plt.subplot(2,1,1)           #changing values of the matrix dimensions
plt.plot(x,y1,color='g',linestyle =':',linewidth = 2)
plt.subplot(2,1,2)
plt.plot(x,y2,color ='r',linestyle = '-')
plt.grid(True)
plt.show()
```

### Bar plot

```python
student = {"Bob":20,"Noah":67,"henry":89}   #creating a dictionary
```

```python
names = list(student.keys())                #converting the keys and their values into lists
marks = list(student.values())
```

```python
plt.bar(names,marks)                      #first is the categories and second is their values
plt.title("Bar plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

### Horizontal Bar Plot

This is similar to bar plot just we use .barh here

```python
plt.barh(names,marks,color = 'r')                      #first is the categories and second is their values
plt.title("Bar plot")
plt.ylabel("Students")
plt.xlabel("Marks")
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
a = [1,2,3,4,5,6,7,8,9]
b = [12,11,13,18,23,56,2,88,19]         #make sure they have equal number of elements

plt.scatter(a,b)
plt.show()
```

There are many attributes here to change the aesthetics and also can have subplots in the same way as created earlier.

### Histogram

```python
data = [1,2,2,2,2,5,5,5,6,6,7,8,9,9,9,9]
plt.hist(data)
plt.ylabel("frequency")
plt.title("Histogram")
plt.show()
```

```python
plt.hist(data,color ='g', bins = 5)        #can change the number of bins
plt.show()
```

### Box - Plot

```python
a = [1,2,3,4,5,6]
b = [11,22,33,44,55,66,77,88]
c = [3,56,78,34,55,90]

data = list([a,b,c])
plt.boxplot(data)
plt.show()
```

### Violin Plot

```python
plt.violinplot(data)
plt.show()
```

```python
plt.violinplot(data,showmedians = True)      #to be able to see median values also
plt.show()
```

### Pie Chart

```python
clubs = ["Music","Fine Arts","Dance","Drama","Sports"]
members = [10,20,50,15,15]

plt.pie(members,labels = clubs)
plt.show()
```

```python
plt.pie(members,labels = clubs,autopct = '%0.2f%%',colors = ["blue","red","green","pink","orange"])
plt.show()
```

### DoughNut Chart

There isn't any unique method for this. We will be using 2 pie charts to create a doughnut chart

```python
plt.pie(members,labels = clubs,radius = 2)
plt.pie([1],colors =['w'],radius = 1)
plt.show()
```
