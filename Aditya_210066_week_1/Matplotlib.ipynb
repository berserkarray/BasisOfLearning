# Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive
visualizations in Python. It offers a viable open source alternative to MATLAB. Most 
of the Matplotlib utilities lies under the pyplot submodule, and are usually imported 
under the plt alias. It is a cross-platform library for making 2D plots from data in 
arrays.


```python
from matplotlib import pyplot as plt 
import numpy as np
```

## Plotting graphs and using different functions in matplotlib

```python
x = [1,4,7,9]
y = [4,7,30,23]
plt.xlabel('X-axis') # labelling x coordinate
plt.ylabel('Y-axis') # labelling y coordinate
plt.plot(x,y)
plt.title('x vs y') # title of graph
plt.show() # Shows the graph
```

```python
v=np.arange(0,15,2)
u=v**2
plt.plot(v,u)
print(v,u)  # prints the values of v & u in the form of array
plt.plot(v,u,marker="o") # to marks the points on graph
plt.show()
```

```python
x = [1,4,7,9,14]
y = [4,7,30,23,12]
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y,marker="x",ms="15")
plt.plot(v,u,marker="o")
plt.show()
```

### Using arrays

```python
i=np.arange(0,15,2)
j=np.arange(4,40,5)
plt.scatter(i,j,label='1')
v=np.arange(0,15,2)
u=v**2
print(v,u)
plt.scatter(v,u,label='2')
plt.title("Sample")
plt.legend()
plt.show()
```

### To change the style of graph

```python
from matplotlib import style
```

```python
v=np.arange(0,15,2)
u=v**2
plt.plot(v,u)
print(v,u) 
plt.plot(v,u,marker="o") 
style.use('fivethirtyeight')
plt.show()
```

```python
x = [1,4,7,9,14]
y = [4,7,30,23,12]
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y,marker="x",ms="15")
style.use('dark_background')
plt.show()
```
